```python
import pandas as pd
pd.set_option('display.max_rows', 500)
```

```python
por_vs_po = pd.read_excel("po_pr_line_differene.xlsx",sheet_name="POR vs PO ")
po_line_no = pd.read_excel("po_pr_line_differene.xlsx",sheet_name="PO Line No")
por_data = pd.read_excel("po_pr_line_differene.xlsx",sheet_name="POR vs PO line item difference")
```

```python
po_data = po_line_no['EBELN'].value_counts().reset_index()
```

```python
po_data.columns = ['PO Number','count']
```

```python
po_data.head(2)
```

```python
por_line_data = por_data['DOCID'].value_counts().reset_index()
```

```python
por_line_data.columns = ['docId','count']
```

```python
por_po_lines = por_vs_po.merge(po_data, how='left')
```

```python
por_po_lines = por_po_lines.fillna(0)
```

```python
por_po_lines['count'] = por_po_lines['count'].astype(int)
```

```python
por_po_lines.head(20)
```

```python
por_line_data.loc[por_line_data['docId']==11000015]
```

```python
new_data = []
for index, row in por_line_data.iterrows():
    docId = row['docId']
    lcount = row['count']
    result = por_po_lines.loc[por_po_lines['POR Ref']==docId]
    pdocid = result['POR Ref'].values[0]
    ppon = result['PO Number'].values[0]
    pcount = result['count'].values[0]
    if lcount!=pcount:
        new_data.append([pdocid,ppon,pcount,lcount])
#     print()
#     if(lcount!=result[0]):
#         print("Por Count {} | Po count {}".format(lcount,result))
```

```python
missing_data = pd.DataFrame(new_data, columns=['por','po','po_lines','por_lines'])
```
